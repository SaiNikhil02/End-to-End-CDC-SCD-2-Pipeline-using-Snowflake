```python
def create_csv_file() -> None:
    current_time = datetime.now().strftime("%Y%m%d_%H%M%S")
    data = defaultdict(list)
    for i in range(RECORD_COUNT):
        data['customer_id'].append(i+1) 
        data['first_name'].append(fake.first_name())
        data['last_name'].append(fake.last_name())
        data['email'].append(fake.email())
        data['street'].append(fake.street_address()) 
        data['city'].append(fake.city()) 
        data["state"].append(fake.state())
        data["country"].append(fake.country())
    df = pd.DataFrame(data) 
    df.to_csv(pwd_path/ f'customer_{current_time}.csv', index = False)
        
            
            
        
        
```

```python
if __name__ == '__main__':
    create_csv_file()
```

```python
import websocket
import json

# Define the stream URL for BTCUSDT trades
stream_url = 'wss://stream.binance.com:9443/stream?streams= btcusdt@trade/\
ethusdt@trade/\
btcusdt@kline_1m/\
ethusdt@kline_1m/\
btcusdt@bookTicker/\
ethusdt@bookTicker'


def on_message(ws, message):
    data = json.loads(message)
    print(f"Trade received: {data}")

def on_error(ws, error):
    print(f"Error: {error}")

def on_close(ws, close_status_code, close_msg):
    print("### Connection Closed ###")

def on_open(ws):
    print("Connection Opened")
    # For raw streams, you just connect; no subscription message needed.

if __name__ == "__main__":
    ws = websocket.WebSocketApp(stream_url,
                                on_message=on_message,
                                on_error=on_error,
                                on_close=on_close)
    ws.on_open = on_open
    ws.run_forever() # Runs indefinitely, handling messages
```
